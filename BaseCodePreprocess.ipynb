```python
import re
import string
puncts = [re.escape(c) for c in string.punctuation]
PUNCTUATION_REGEX = re.compile('|'.join(puncts))
def replace_percentage(text):
    return re.sub(r'%', ' phần trăm ', text)
def remove_punctuation(text):
    return re.sub(PUNCTUATION_REGEX, '', text)
def replace_special_words(text):
    # replace special words
    replace_list = {
        'ô kêi': ' ok ', 'o kê': ' ok ',
        'kh ':' không ', 'kô ':' không ', 'hok ':' không ',
        'kp ': ' không phải ', 'kô ': ' không ', 'ko ': ' không ', 'khong ': ' không ', 'hok ': ' không ',
        'luông':'luôn','hcoj':'học'
    }
    for k, v in replace_list.items():
      text = text.replace(k, v)
    return text
def remove_duplicate_space(text):
    return re.sub(r'\s+', ' ', text)
def remove_duplicate(text):
    text = re.sub(r'\b(\w+)\s+\1\b', r'\1', text)
    return text
def preprocessing(text):
    text = text.lower()
    text = process_above(text)
    text = remove_duplicate_space(text)
    text = replace_percentage(text)
    text = replace_special_words(text)
    text = remove_duplicate(text)
    text=remove_punctuation(text)
    return text.replace("\n","")
def process_above(text):
    text_search=re.compile('[0-9]+\/[0-9]+').search(text)
    if text_search:
      print(text_search)
    return text
```
